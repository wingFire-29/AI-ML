```python
oscar_movies=[
    "The dark knight rises"
    "The Heart Locker"
    "The King's speech"
    "The Artist"
    "Argo"
]

years=[2008,2009,2010,2011,2012]
revenue=[1005,282,271,170,427] #in $M

#plot
plt.plot(years,revenue)
plt.title("Oscar Movies Year-wise revenue")
plt.xlabel("years")
plt.xlabel("Revenue in M$")
```

```python
#Non oscar movies

non_oscar_Movies=["Slumdof millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue=[378,2788,828,198,275]

plt.plot(years,revenue,color="red",marker="v",linestyle="--",linewidth=5,label="Oscar Movies")
plt.plot(years,non_oscar_revenue,"o-b",label="Non Oscar Movies")

plt.title("Oscar Movies vs Non Oscar Movies")
plt.xlabel("year")
plt.ylabel("Revenue in M$")
          
plt.legend()
```
