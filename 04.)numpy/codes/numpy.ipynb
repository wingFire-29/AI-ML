```python
import numpy as np
import time
```

```python
# execution performance
# List VS numpy arrays

size=10_000_000
#python list
py_list=list(range(size))
start=time.time()
sq_list=[x**2 for x in py_list]
end=time.time()
print(f"python list time={end-start} seconds")

#numpy arrays
np_array=np.array(py_list)
start=time.time()
sq_array=np_array ** 2
end=time.time()
print(f"numpy array time={end-start} seconds")
```

```python
#memory
import sys
print(f"python list size={sys.getsizeof(py_list) * len(py_list)} bytes")
print(f"numpy array size={np_array.nbytes} bytes")
```

```python
#copy vs view
# python list creates a copy of an array/list 
# while np array creates a view of an array


# list

# nums=[1,2,3,4,5]
# sub_list=nums[1:3]
# print(sub_list)
# sub_list[0]=200
# print(sub_list)
# print(nums)

# np array
arr=np.array([1,2,3,4,5])
sub_arr=arr[1:3].copy()
print(sub_arr)
sub_arr[0]=200
print(sub_arr)
print(arr)
#this is a like a reference pointer but to actually make a copy we can use ".copy" function
```
