```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#fetching data

page_count=1
while True:
    URL=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(URL)

    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")

    if not quotes:
        print("No valid pages anymore")
        break
    with open(f"Scraped_data/quotes{page_count}.html","w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"Data Downloaded from {page_count} page successfully ")
    page_count=page_count+1    
```

```python
#Extracting the useful info -page1

with open("Scraped_data/quotes1.html","r",encoding="utf-8") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content,"lxml")
```

```python

page=1
life_quotes=[]
while page<=10:
    
    with open(f"Scraped_data/quotes{page}.html","r",encoding="utf-8") as f:
        html_content=f.read()
        soup=BeautifulSoup(html_content,"lxml")
        all_quotes=soup.select("div.quote")
        
        for q in all_quotes:
            all_tags=[]
            quotes={}
            for tag in q.select(".tags .tag"):
                
                all_tags.append(tag.get_text())   
        
            if "life" in all_tags:
                text=q.select_one("span.text").get_text()
                author=q.select_one("small.author").get_text()
                quotes["Quote"]=text
                quotes["author"]=author
                life_quotes.append(quotes)  
    page=page+1
# print(len(life_quotes))    
# print(life_quotes[:3])    
df=pd.DataFrame(life_quotes)
```

```python
df.to_csv("Scraped_data/life_quotes.csv",index=False)
```
